```python
import random as rnd

def answer():
    flip = rnd.uniform(0,1)
    ans = False
    if flip <= 0.5:
        ans = True
    else:
        flip = rnd.uniform(0,1)
        if flip > 0.5:
            ans = True
            
    return ans
```

```python
n = 50

trues = 0
for i in range(n):
    if answer() == True:
        trues += 1
    
print(trues/n)
```
